```python
import requests
```

```python
inp = input()
respon = requests.get('https://in.search.yahoo.com/search?p={}'.format(inp))
```

```python
from bs4 import BeautifulSoup
bs = BeautifulSoup(respon.text,'html.parser')
```

```python
print(bs.prettify())
```

```python
links = bs.find_all('a')
```

```python
for i in links:
    x = i.get('class')
    if(x != None and x[0] == "ac-algo"):
        print(i.get('href'))
```
