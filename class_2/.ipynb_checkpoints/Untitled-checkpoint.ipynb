```python
inp1 = input()
resp = requests.get("https://en.wikipedia.org/wiki/{}".format(inp1))
from bs4 import BeautifulSoup
soup = BeautifulSoup(resp.text, 'html.parser')
p_tag_list = soup.find_all("p")
for i in p_tag_list:
    if(i.get("class")==None):
        print(i.text)
```

```python
class car:
    def __init__(self,engineno,color):
        self.car_color = color
        self.car_engine = engineno
        self.distance_covered = 0
        
    def total_kitna_chali(self):
        return self.distance_covered
    
    def abhi_kitna_chali(self,x):
        self.distance_covered = self.distance_covered+x
        
    
```

```python
car1  = car('01234','Red')
```

```python
car1.total_kitna_chali()
```

```python
car1.abhi_kitna_chali(100)
```

```python
car1.total_kitna_chali()
```
